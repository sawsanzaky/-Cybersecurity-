 **Error Handling & Debugging  Practical Python**
 Goal

Write robust, fault-tolerant, and debuggable Python code.

```python
# ------------------------------
# 1 Syntax vs Runtime Errors
# Syntax Error (uncomment to see)
# print("Hello World"   #  Missing closing parenthesis

# Runtime Error
try:
    num = int(input("Enter a number: "))
    print("Number divided by 2:", num / 2)
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input! Please enter a number.")
print("-" * 40)

# ------------------------------
# 2 try / except / else / finally
try:
    n = int(input("Enter index to access: "))
    lst = [10, 20, 30]
    value = lst[n]
except IndexError:
    print("Index out of range!")
else:
    print("Accessed value:", value)
finally:
    print("Execution finished (always runs)")
print("-" * 40)

# ------------------------------
# 3 Custom Exceptions
class InvalidPortError(Exception):
    pass

port = int(input("Enter port number: "))

try:
    if port < 0 or port > 65535:
        raise InvalidPortError("Port number must be 0-65535!")
    print(f"Port {port} is valid")
except InvalidPortError as e:
    print("Custom Exception:", e)
print("-" * 40)

# ------------------------------
# 4 Debugging Techniques
# Use print statements to trace code
def debug_example():
    for i in range(3):
        print(f"Debug: i={i}")  # trace variable
        result = 10 / (i - 1)   # will cause ZeroDivisionError
        print(result)

try:
    debug_example()
except ZeroDivisionError as e:
    print("Caught Error during debugging:", e)
print("-" * 40)

# ------------------------------
# 5 Logging Basics
import logging

logging.basicConfig(level=logging.INFO)
logging.info("Script started")

try:
    x = int(input("Enter number to divide 100 by: "))
    result = 100 / x
except ZeroDivisionError:
    logging.error("Division by zero attempted")
else:
    logging.info(f"Division successful: {result}")
finally:
    logging.info("Script finished")
```

```python
# ------------------------------
# 1 List & Dictionary Comprehensions
# List comprehension: squares of even ports
ports = [22, 80, 443, 8080, 21]
even_squares = [p**2 for p in ports if p % 2 == 0]
print("Even port squares:", even_squares)

# Dictionary comprehension: port status
port_status = {p: "Open" if p in [22, 80, 443] else "Closed" for p in ports}
print("Port status dict:", port_status)
print("-" * 40)

# ------------------------------
# 2 Generators & Iterators
# Generator: yield port numbers
def port_generator(ports):
    for p in ports:
        yield p

for p in port_generator(ports):
    print("Scanning port:", p)
print("-" * 40)

# ------------------------------
# 3 Decorators
# Example: log function execution
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"[LOG] Running {func.__name__} with args={args}, kwargs={kwargs}")
        result = func(*args, **kwargs)
        print(f"[LOG] Result: {result}")
        return result
    return wrapper

@logger
def scan_port(port):
    return "Open" if port in [22, 80, 443] else "Closed"

scan_port(22)
scan_port(8080)
print("-" * 40)

# ------------------------------
# 4 Context Managers
# Using with statement to safely open a file
with open("log.txt", "w") as f:
    f.write("Port scan started\n")
    f.write("All ports scanned successfully\n")
print("File operations completed safely")
print("-" * 40)

# ------------------------------
# 5 Multithreading
import threading
import time

def thread_scan(port):
    print(f"Thread scanning port {port}")
    time.sleep(0.5)
    print(f"Thread finished scanning port {port}")

threads = []
for port in [22, 80, 443]:
    t = threading.Thread(target=thread_scan, args=(port,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
print("All threads completed")
print("-" * 40)

# ------------------------------
# 6 Multiprocessing
from multiprocessing import Process

def process_scan(port):
    print(f"Process scanning port {port}")

processes = []
for port in [21, 23]:
    p = Process(target=process_scan, args=(port,))
    processes.append(p)
    p.start()

for p in processes:
    p.join()
print("All processes completed")
print("-" * 40)

# ------------------------------
# 7 Async Programming
import asyncio

async def async_scan(port):
    print(f"Async scanning port {port}")
    await asyncio.sleep(0.5)
    print(f"Async finished scanning port {port}")

async def main():
    tasks = [async_scan(p) for p in [25, 110]]
    await asyncio.gather(*tasks)

asyncio.run(main())
print("Async scanning completed")
```

1 **Password Hashing & Validation**

```python
import hashlib

# Hash password
password = "SecurePass123"
hashed = hashlib.sha256(password.encode()).hexdigest()
print("Hashed password:", hashed)

# Validate input
user_input = input("Enter password: ")
if hashlib.sha256(user_input.encode()).hexdigest() == hashed:
    print(" Access granted")
else:
    print(" Access denied")
```

5 **Network Scanning Scripts**

Definition: Scan network hosts or ports to detect open services (ethical use).

Example: Simple port scanner

```python
import socket

hosts = ["127.0.0.1"]
ports = [22, 80, 443]

for host in hosts:
    for port in ports:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(0.5)
        result = sock.connect_ex((host, port))
        if result == 0:
            print(f"Port {port} on {host} is open")
        else:
            print(f"Port {port} on {host} is closed")
        sock.close()
```
